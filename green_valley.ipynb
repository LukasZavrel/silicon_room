## First Assignment

```python
def mysterious_function(nr: int, nr_steps: int = 0) -> int:
    if nr == 1:
        return nr_steps
    if nr % 2:  # % is modulo operator, e.g. 8 % 3 = 2
        return mysterious_function(3 * nr + 1, nr_steps + 1)
    else:
        return mysterious_function(nr // 2, nr_steps + 1)
```

```python
print(max([mysterious_function(i) for i in range(10000)]))
```
